# The 21 Game

```python
import random

while True:
    
    current_number = 1 # khởi tạo giá trị ban đầu bằng 0 

    if random.randint(0, 1) == 0: # sinh ngẫu nhiên chọn người chơi
        current_player = "human"
    else:
        current_player = "computer"

    while current_number <= 21:
      
        print("Số hiện tại là " + str(current_number) + ".") # in ra giá trị curent_number
        print()

        if current_player == "human":

            print("Thêm 1,2 hoặc 3. Không vượt quá 21. Người chơi vượt quá 21 sẽ thua.")

            player_choice = ""
            while player_choice not in ["1", "2", "3"]:
                player_choice = input("Bạn thêm giá trị là bao nhiêu: ")

            player_choice = int(player_choice)
            current_number = current_number + player_choice
            print()

            if current_number >= 21:
                print("Số hiện tại là " + str(current_number) + ".")
                print()
                print("Sorry, bạn đã thua.")
                break
            else:
                current_player = "computer"

        else:

            computer_choice = random.randint(1, 3)
            print("Lượt của máy. Máy tính chọn " + str(computer_choice) + ".")
            print()
            
            current_number = current_number + computer_choice
            
            if current_number >= 21:
                print("Số hiện tại là:  " + str(current_number) + ".")
                print()
                print("Lam tốt lắm, bạn thắng!")
                break
            else:
                current_player = "human"

    play_again = input("Bạn có muốn chơi lại, nhấn y để đồng ý, nhấn phím bất kỳ để thoát: ")
    if play_again.lower().startswith("y"):
        continue
    else:
        print("Goodbye")
        break
```
